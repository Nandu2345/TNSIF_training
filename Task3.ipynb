```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global YouTube Statistics.csv",encoding='latin-1')
print(df.head())

top_subscribers = df[['Youtuber', 'subscribers']].sort_values(
    by='subscribers', ascending=False).head(10)
print("\nTop 10 Channels by Subscribers:")
print(top_subscribers)

plt.figure(figsize=(10,5))
plt.bar(top_subscribers['Youtuber'], top_subscribers['subscribers'])
plt.xticks(rotation=90)
plt.xlabel("YouTube Channel")
plt.ylabel("Subscribers")
plt.title("Top 10 YouTube Channels by Subscribers")
plt.show()
top_views = df[['Youtuber', 'video views']].sort_values(
    by='video views', ascending=False).head(10)

print("\nTop 10 Channels by Video Views:")
print(top_views)

plt.figure(figsize=(10,5))
plt.bar(top_views['Youtuber'], top_views['video views'])
plt.xticks(rotation=90)
plt.xlabel("YouTube Channel")
plt.ylabel("Video Views")
plt.title("Top 10 YouTube Channels by Video Views")
plt.show()

print("\nTop 10 Channels by Video Views:")
print(top_views)

plt.figure(figsize=(10,5))
plt.bar(top_views['Youtuber'], top_views['video views'])
plt.xticks(rotation=90)
plt.xlabel("YouTube Channel")
plt.ylabel("Video Views")
plt.title("Top 10 YouTube Channels by Video Views")
plt.show()
```
