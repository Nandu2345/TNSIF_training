#Supervised Learning
##1. What is Supervised Learning?

Definition:

Supervised Learning is a type of Machine Learning in which the model learns from labeled data. Every training example contains:

- Input (Features/Independent Variables - X)
- Output (Target/Dependent Variable - Y)

The algorithm learns the relationship between the input and output so that it can predict the output for new, unseen data.

-----------------------------------------------


Formula

Input (X)  ------> Machine Learning Algorithm ------> Output (Y)

---------------------------------------------------------

Simple Example

Suppose we want to predict whether a student will pass or fail.

Study Hours	Result
2	Fail
3	Fail
5	Pass
7	Pass
9	Pass

Here,

Input (X) = Study Hours
Output (Y) = Pass/Fail

The model learns this relationship and predicts the result for a student who studies for 6 hours.

Real-Life Analogy

Imagine a teacher teaching children.

The teacher shows many examples.

Cat  → Cat,
Dog  → Dog,
Cow  → Cow

After seeing many labeled examples, the child can identify a new animal correctly.

Similarly,

Teacher = Training Data,
Student = Machine Learning Model,
Learning = Training Process.

-----------------------------------------------

###Workflow of Supervised Learning

Collect Data

      ↓

Label the Data

      ↓

Split into Training & Testing Data

      ↓
Train the Model

      ↓
Evaluate the Model

      ↓
Predict New Data

----------------------------------------------------


####Advantages
- High prediction accuracy
- Easy to evaluate
- Suitable when labeled data is available
- Widely used in industry
####Disadvantages
- Requires large labeled datasets
- Labeling data is expensive
- Cannot discover hidden patterns without labels

Supervised Learning is divided into two major types:

  1. Regression

##Simple Linear Regression (Regression)
What is it?

Simple Linear Regression is used to predict a continuous value using one independent variable.

Formula -->
y=mx+c

Where

x → Independent Variable,
y → Dependent Variable,
m → Slope,
c → Intercept


```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Dataset
df = pd.DataFrame({
    "Experience":[1,2,3,4,5],
    "Salary":[3,4,5,6,7]
})

# Independent Variable
X = df[["Experience"]]

# Dependent Variable
y = df["Salary"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X,y)

# Prediction
prediction = model.predict([[6]])

print("Predicted Salary:",prediction[0])

print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
```

##Multiple Linear Regression

What is it?

Multiple Linear Regression predicts the output using more than one input feature.

Formula
Y=b0+b1X1+b2X2+...+bnXn
	​

Example

Predict House Price

Inputs

Area,
Bedrooms,
Age,

Output

House Price

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({
    "Area":[1000,1200,1500,1800,2000],
    "Bedrooms":[2,2,3,3,4],
    "Age":[10,8,5,2,1],
    "Price":[40,45,60,75,90]
})

X = df[["Area","Bedrooms","Age"]]
y = df["Price"]

model = LinearRegression()

model.fit(X,y)

prediction = model.predict([[1700,3,4]])

print("Predicted Price:",prediction[0])
```

Decision Tree
What is it?

A Decision Tree makes decisions using a sequence of if-else conditions, similar to a flowchart.

Example

Loan Approval

Income > 50,000?

        Yes
         |
 Credit Score >700?

     Yes      No
      |        |
 Approve    Reject

No
|
Reject

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df = pd.DataFrame({
    "Income":[25,35,45,60,70,80],
    "Loan":[0,0,0,1,1,1]
})

X = df[["Income"]]
y = df["Loan"]

model = DecisionTreeClassifier()

model.fit(X,y)

prediction = model.predict([[65]])

print("Loan Approved:",prediction[0])
```

##Logistic Regression (Classification)
What is Logistic Regression?

Logistic Regression is a Supervised Machine Learning Classification algorithm used to predict categorical outcomes, especially Binary Classification.

Although its name contains "Regression", it is used for Classification, not Regression.


#### Why is it called Logistic Regression?

Unlike Linear Regression, which predicts continuous values, Logistic Regression predicts the probability of belonging to a class.

It uses the Sigmoid Function to convert any value into a probability between 0 and 1.

```python

# Step 1: Import Libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Step 2: Create Dataset
df = pd.DataFrame({
    "Hours":[1,2,3,4,5,6],
    "Pass":[0,0,0,1,1,1]
})

print("Dataset")
print(df)

# Step 3: Input and Output
X = df[["Hours"]]
y = df["Pass"]
# Step 4: Create Model
model = LogisticRegression()
# Step 5: Train Model
model.fit(X,y)
# Step 6: Prediction
prediction = model.predict([[3.5]])
# Step 7: Output
print("\nPrediction:",prediction[0])

if prediction[0]==1:
    print("Student Passed")
else:
    print("Student Failed")
```

##Naive Bayes Classification
What is Naive Bayes?

Naive Bayes is a Supervised Machine Learning Classification algorithm based on Bayes' Theorem.

It predicts the class having the highest probability.

It is widely used for

Spam Detection,
Sentiment Analysis,
Email Classification,
Medical Diagnosis.

Bayes' Theorem P(A∣B)=P(B)P(B∣A)/P(A)
- P(A|B) → Probability of A given B
- P(B|A) → Probability of B given A
- P(A) → Prior Probability
- P(B) → Evidence

The algorithm calculates the probability for each class and predicts the class with the highest probability.
	​


Real-Life Example

Suppose an email contains these words:

Congratulations!

You Won Free Prize

Naive Bayes calculates

Probability(Spam) = 0.95

Probability(Not Spam) = 0.05

```python

# Step 1: Import Libraries
import pandas as pd
from sklearn.naive_bayes import GaussianNB

# Step 2: Create Dataset
df = pd.DataFrame({
    "Marks":[35,40,45,60,70,80],
    "Pass":[0,0,0,1,1,1]
})

print("Dataset")
print(df)

# Step 3: Input and Output
X = df[["Marks"]]
y = df["Pass"]

# Step 4: Create Model
model = GaussianNB()

# Step 5: Train Model
model.fit(X,y)

# Step 6: Prediction
prediction = model.predict([[65]])

# Step 7: Output
print("\nPrediction:",prediction[0])

if prediction[0]==1:
    print("Student Passed")
else:
    print("Student Failed")
```
