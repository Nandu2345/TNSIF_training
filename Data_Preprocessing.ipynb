##  What is Data Cleaning?

Data Cleaning is the process of:
- Fixing incorrect data
- Handling missing values
- Removing duplicates
- Standardizing formats

---

##  Why is Data Preprocessing Important?

Machine Learning models:
- Cannot handle missing or messy data properly
- Require numerical and consistent input

---

##  Example of Messy Data:

| Name | Age | City     |
|------|-----|----------|
| Ram  | 25  | Mumbai   |
| Shyam| NaN | mumbai   |
| Ram  | 25  | Mumbai   |

Problems:
- Missing values
- Inconsistent text
- Duplicate rows

```python
data = {
    "Name": ["Ram", "Shyam", "Ram", "Sita", None,"Shyam","Shyam","Shyam","Ram"],
    "Age": [25, np.nan, 25, 30, 22,13,17,18,19],
    "City": ["Mumbai", "mumbai", "Mumbai", "Delhi", "Delhi", "Delhi", "Delhi", "Delhi", "Delhi"]
}

df = pd.DataFrame(data)
df
```

##  Common Data Issues

1. Missing Values → NaN
2. Duplicate Records
3. Incorrect Data Types
4. Inconsistent Text (Mumbai vs mumbai)

---

We must identify and fix these issues before using data.

```python
# First few rows
print(df.head())
print('this is info()====================')

# Data info
print(df.info())
print('this is describe()====================')
# Summary statistics
print(df.describe())
```

##  Missing Values

Missing values are represented as:
- NaN (Not a Number)
- None

---

##  Steps:
1. Identify missing values
2. Count them
3. Decide:
   - Drop
   - Fill

```python
# Check missing values
print(df.isnull())

# Count missing values
df.isnull().sum() #count of missing values
```

```python
# Drop rows with missing values
df_drop = df.dropna()
df_drop
```

##  Filling Missing Values

### Numerical Columns:
- Mean
- Median

### Categorical Columns:
- Mode

---

##  When to Use:
- Mean → Normal data
- Median → Skewed data
- Mode → Categories

```python
#  Best Practice

df["Age"] = df["Age"].fillna(df["Age"].mean())  #targeting the column
df["Name"] = df["Name"].fillna(df["Name"].mode()[0])
#we dont have any missing values in city column

df
```

##  Selecting Multiple Columns

To work on multiple columns together:

###  Syntax:
df[["col1", "col2"]]

---

##  Why use this?
- Apply same operation on multiple columns
- Clean multiple columns at once
- Save time and avoid repetition

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Age": [20, np.nan, 25, 30,20],
    "Salary": [20000, 25000, np.nan, 40000,20000],
    "City": ["Mumbai", "Delhi", "Pune", "Mumbai","Mumbai"]
})

df
```

```python
# Fill multiple columns with mean

df[["Age", "Salary"]] = df[["Age", "Salary"]].fillna(df[["Age", "Salary"]].mean())

df
```

```python
# Convert multiple columns to integer

df[["Age", "Salary"]] = df[["Age", "Salary"]].astype(int)

df.dtypes
df
```

```python
# Increase values by 10%

df[["Age", "Salary"]] = df[["Age", "Salary"]].apply(lambda x: x * 1.1)

df
```

```python
df
```

```python
# Replace specific values

df[["City"]] = df[["City"]].replace("Mumbai", "MUMBAI")

df
```

```python
df
```

## 🔹 Encoding (Categorical → Numerical)

Machine Learning models only understand numbers.

---

### Types of Data:
1. Nominal → No order (City, Color)
2. Ordinal → Ordered (Low, Medium, High)

---

### Methods:
- One-Hot Encoding
- Label Encoding

```python
df_encoded = pd.get_dummies(df, columns=["City"])
df_encoded
```

Converts categories into numbers:

Example:
Mumbai → 0  
Delhi → 1  

⚠️ Not ideal for nominal data (creates fake order)

## 🔹 Convert Date Column

Convert string → datetime format

---

Helps in:
- Sorting
- Time-based analysis

```python
df["date"] = df["date"].astype(str).str.strip()
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df.info()
```

## 🔹 Handle Missing Values

### Numerical Columns:
- rainfall
- temperature
- humidity
- wind_speed

➡️ Fill with mean

### Categorical:
- weather_condition

➡️ Fill with mode

```python
import pandas as pd

# Numerical columns
num_cols = ["rainfall", "temperature", "humidity", "wind_speed"]

# Fill numerical missing values with mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill categorical missing values with mode
df["weather_condition"] = df["weather_condition"].fillna(df["weather_condition"].mode()[0])

# Check missing values
print(df.isnull().sum())
```

## 🔹 Fix Text Inconsistency

Standardize text:
- Rainy → rainy
- SUNNY → sunny

```python
df["weather_condition"] = df["weather_condition"].str.lower()

df["weather_condition"].unique()

df
```

# 🎓 Student Performance Dataset – Data Cleaning

## 🎯 Objective:
Clean and preprocess student dataset for ML

---

## 🔹 Columns:
- name
- age
- gender
- marks
- attendance
- city

---

## ⚠️ Possible Issues:
- Missing values
- Inconsistent text (Male / male)
- Invalid marks (>100 or <0)
- Duplicates

```python
import pandas as pd
import numpy as np

data = {
    "name": ["Amit", "Sara", "John", "Amit", None],
    "age": [20, 21, np.nan, 20, 19],
    "gender": ["Male", "female", "Male", "Male", "FEMALE"],
    "marks": [85, 90, 105, 85, -5],
    "attendance": [90, 85, 80, 90, None],
    "city": ["Mumbai", "Delhi", "mumbai", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
df.isnull().sum()
df.describe()
```

## 🔹 Handling Missing Values

- age → mean
- attendance → mean
- name → "Unknown"

```python
df["age"].fillna(df["age"].mean(), inplace=True)
df["attendance"].fillna(df["attendance"].mean(), inplace=True)
df["name"].fillna("Unknown", inplace=True)

df
```

```python
df["gender"] = df["gender"].str.lower()
df["city"] = df["city"].str.lower()

df
```

## 🔹 Fix Invalid Data

Marks should be:
- Between 0 and 100

```python
df = df[(df["marks"] >= 0) & (df["marks"] <= 100)]

df
```
