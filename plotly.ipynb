###What is Plotly?

Plotly is a Python library used to create interactive and attractive visualizations.

Unlike Matplotlib and Seaborn, Plotly allows users to:

Zoom in and out,
Hover over data points,
Pan across the graph,
Save graphs as images,
Interact with charts directly.

Create Dataset


Import Plotly Express

      
Choose a Chart (Bar, Pie, Scatter...)

      
Specify Columns for Axes

      
Add Title (Optional)

      
Display Chart using fig.show()

```python
import pandas as pd
import plotly.express as px
```

```python
import pandas as pd

data = {
    "Department": ["IT", "HR", "Finance", "Marketing"],
    "Employees": [50, 20, 30, 25]
}

df = pd.DataFrame(data)

df
```

```python
import plotly.express as px

fig = px.bar(
    df,
    x="Department",
    y="Employees",
    title="Number of Employees in Each Department"
)

fig.show()
```

```python
fig = px.pie(
    df,
    names="Department",
    values="Employees",
    title="Employee Distribution by Department"
)

fig.show()
```

```python
data = {
    "Experience": [1, 2, 3, 4, 5 ,3 ,4 ,6],
    "Salary": [30000, 40000, 50000, 65000, 80000, 62000, 35000, 23000]
}

df = pd.DataFrame(data)

fig = px.scatter(
    df,
    x="Experience",
    y="Salary",
    title="Experience vs Salary"
)

fig.show()
```

```python
fig = px.histogram(
    df,
    x="Salary",
    title="Salary Distribution"
)

fig.show()
```
